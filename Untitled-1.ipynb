```python
import asyncio
import subprocess

async def run_command(cmd, check=True, capture_output=False):
    global current_process
    program = cmd[0]
    args = cmd[1:]
    stdout_pipe = asyncio.subprocess.PIPE if capture_output else None
    stderr_pipe = asyncio.subprocess.PIPE if capture_output else None
    process = await asyncio.create_subprocess_exec(program, *args, stdout=stdout_pipe, stderr=stderr_pipe)
    set_current_process(process, "unknown", "subprocess")
    try:
        stdout, stderr = await process.communicate()
        if check and process.returncode != 0:
            raise subprocess.CalledProcessError(process.returncode, cmd, output=stdout, stderr=stderr)
        return subprocess.CompletedProcess(cmd, process.returncode, stdout, stderr)
    finally:
        clear_current_process()
```
